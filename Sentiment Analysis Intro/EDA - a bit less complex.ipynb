```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
con = sqlite3.connect('database.sqlite')
data = pd.read_sql_query('select * from Reviews', con)
con.close()
```

```python
display(data.sample(20))
display(data.head(20))
display(data.tail(20))
```

```python
#check for missing values
print(data.isnull().sum())
print('\n\n\n')

def checkweirddataandnull(name):
    print('='*30)
    print('Column name: '+name)
    if data[name].dtypes == 'O':
        display(data[(data[name] == 'nan')|(data[name] == 'NaN')|
                     (data[name] == 'N/A')|(data[name] == '0')|
                     (data[name] == '')|(data[name] == '-1')|
                     (data[name] == 'null')|(data[name] == 'Null')|
                     (data[name] == 'NA')|(data[name] == 'na')|
                     (data[name] == 'none')|(data[name] == 'unknown')])
    else:
        display(data[data[name] < 0])

for i in data.columns:
    checkweirddataandnull(i)
    
```

```python
data['ProfileName'] = data['ProfileName'].apply(lambda x: 'Anonymous' if 
                                                (x == 'nan')|(x == 'NaN')|
                                                (x == 'N/A')|(x == '0')|
                                                (x == '')|(x == '-1')|
                                                (x == 'null')|(x == 'Null')|
                                                (x == 'NA')|(x == 'na')|
                                                (x == 'none')|(x == 'unknown') else x)
data['Summary'] = data['Summary'].apply(lambda x: 'No Summary' if 
                                        (x == 'nan')|(x == 'NaN')|
                                        (x == 'N/A')|(x == '0')|
                                        (x == '')|(x == '-1')|
                                        (x == 'null')|(x == 'Null')|
                                        (x == 'NA')|(x == 'na')|
                                        (x == 'none')|(x == 'unknown') else x)
```

We would like to do a sentiment analysis, which means to understand which comments are supposedly positive and which comments are negative. Lucky for us, we have a score column.

```python
#check for score column distribution
sns.countplot(x = 'Score',data = data)
plt.title('Sentiment Distribution')
plt.show()
```

We can see from the plot above that the majority of user give score 5/5. The probability distributions it self is a little bit skewed to the left. We can say that users that gave a score 4/5 are users who's neutral. Users who gave score below 4 are users who's dissatisfied.

```python
#See what is the most used words from our data
from wordcloud import WordCloud,STOPWORDS
stpwrd = set(STOPWORDS)
stpwrd.update(['br','href']) #this is to handle should there be any HTML tags
text = ' '.join(i for i in data.Text) #this is to combine all text into one huge string.
wordcloud = WordCloud(stopwords = stpwrd, background_color='white', random_state = 42).generate(text)

plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

We can see from the wordcloud above, most of the reviews have been relatively positive if we because we see 'Love' and 'Good' appear on the wordcloud. This is if we assume our customers weren't sarcastic :)

our assumption for most of the reviews being positive is also supported by the count of 5/5 votes being relatively high.

## Classifying Sentiments

Now, we would want to see how the reviews look like if we classify customers by their score. As stated before, 5/5 score will be classified as 'good', 4/5 will be classified as neutral, below 3 score will be classified as 'bad'

```python
data['Sentiments'] = data['Score'].apply(lambda x: 1 if x == 5 else 0 if x == 4 else -1)
```

```python
print(data.Score.value_counts())
print(data.Sentiments.value_counts())
```

## Wordcloud by Sentiments

we update the stopword with the word "Amazon", "Product", "one", and "find" since it shows no meaning.

```python
stpwrd.update(['Amazon','product','one','find'])
```

```python
postext = ' '.join(i for i in data[data.Sentiments == 1].Text) #this is to combine all text into one huge string.
poswordcloud = WordCloud(width = 1600,height = 800, stopwords = stpwrd, background_color='white', random_state = 42).generate(postext)

fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize = (25,25))
ax1.imshow(poswordcloud, interpolation = 'bilinear')
ax1.axis('off')
ax1.set_title('Positive Reviews Wordcloud')

neutext = ' '.join(i for i in data[data.Sentiments == 0].Text) #this is to combine all text into one huge string.
neuwordcloud = WordCloud(width = 1600,height = 800,stopwords = stpwrd, background_color='white', random_state = 42).generate(neutext)

ax2.imshow(neuwordcloud, interpolation = 'bilinear')
ax2.axis('off')
ax2.set_title('Neutral Reviews Wordcloud')

negtext = ' '.join(i for i in data[data.Sentiments == -1].Text) #this is to combine all text into one huge string.
negwordcloud = WordCloud(width = 1600,height = 800,stopwords = stpwrd, background_color='white', random_state = 42).generate(negtext)

ax3.imshow(negwordcloud, interpolation = 'bilinear')
ax3.axis('off')
ax3.set_title('Negative Reviews Wordcloud')

plt.show()
```

We're having a hard time differentiating each sentiments since "taste", "flavor", and "good" are commonly used in each sentiments. We should update the stopwords with these words since this means that those words are quite redundant in differentiating review sentiments.

```python
stpwrd.update(['taste','flavor','good'])
```

```python
postext = ' '.join(i for i in data[data.Sentiments == 1].Text) #this is to combine all text into one huge string.
poswordcloud = WordCloud(width = 1600,height = 800, stopwords = stpwrd, background_color='white', random_state = 42).generate(postext)

fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize = (25,25))
ax1.imshow(poswordcloud, interpolation = 'bilinear')
ax1.axis('off')
ax1.set_title('Positive Reviews Wordcloud')

neutext = ' '.join(i for i in data[data.Sentiments == 0].Text) #this is to combine all text into one huge string.
neuwordcloud = WordCloud(width = 1600,height = 800,stopwords = stpwrd, background_color='white', random_state = 42).generate(neutext)

ax2.imshow(neuwordcloud, interpolation = 'bilinear')
ax2.axis('off')
ax2.set_title('Neutral Reviews Wordcloud')

negtext = ' '.join(i for i in data[data.Sentiments == -1].Text) #this is to combine all text into one huge string.
negwordcloud = WordCloud(width = 1600,height = 800,stopwords = stpwrd, background_color='white', random_state = 42).generate(negtext)

ax3.imshow(negwordcloud, interpolation = 'bilinear')
ax3.axis('off')
ax3.set_title('Negative Reviews Wordcloud')

plt.show()
```

We can see that we can differentiate positive sentiments and negative sentiments since on negative sentiments there are no "love" but we cannot differentiate positive sentiments and neutral sentiments. This means our previous assumption is false. We should re-classify our sentiments into two sentiments only and we should update our stopwords with "buy", "make", and "coffee".

```python
#the idea is to turn 0(neutral sentiments) into 1(positive sentiment)
data['Sentiments'] = data['Sentiments'].apply(lambda x: 1 if x == 0  else 0 if x == -1 else x)
print(data.Sentiments.value_counts())
```

```python
stpwrd.update(['buy','make','coffee'])
```

```python
postextnew = ' '.join(i for i in data[data.Sentiments == 1].Text) #this is to combine all text into one huge string.
poswordcloudnew = WordCloud(width = 1600,height = 800, stopwords = stpwrd, background_color='white', random_state = 42).generate(postextnew)

fig, (ax1, ax2) = plt.subplots(1,2,figsize = (25,25))
ax1.imshow(poswordcloudnew, interpolation = 'bilinear')
ax1.axis('off')
ax1.set_title('Positive Reviews Wordcloud')

negtextnew = ' '.join(i for i in data[data.Sentiments == 0].Text) #this is to combine all text into one huge string.
negwordcloudnew = WordCloud(width = 1600,height = 800,stopwords = stpwrd, background_color='white', random_state = 42).generate(negtextnew)

ax2.imshow(negwordcloudnew, interpolation = 'bilinear')
ax2.axis('off')
ax2.set_title('Negative Reviews Wordcloud')

plt.show()
```
