```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
con = sqlite3.connect('database.sqlite')
data = pd.read_sql_query('select * from Reviews', con)
con.close()
```

```python
display(data.sample(20))
display(data.head(20))
display(data.tail(20))
```

```python
#check for missing values
print(data.isnull().sum())
print('\n\n\n')

def checkweirddataandnull(name):
    print('='*30)
    print('Column name: '+name)
    if data[name].dtypes == 'O':
        display(data[(data[name] == 'nan')|(data[name] == 'NaN')|
                     (data[name] == 'N/A')|(data[name] == '0')|
                     (data[name] == '')|(data[name] == '-1')|
                     (data[name] == 'null')|(data[name] == 'Null')|
                     (data[name] == 'NA')|(data[name] == 'na')|
                     (data[name] == 'none')|(data[name] == 'unknown')])
    else:
        display(data[data[name] < 0])

for i in data.columns:
    checkweirddataandnull(i)
    
```

```python
data['ProfileName'] = data['ProfileName'].apply(lambda x: 'Anonymous' if 
                                                (x == 'nan')|(x == 'NaN')|
                                                (x == 'N/A')|(x == '0')|
                                                (x == '')|(x == '-1')|
                                                (x == 'null')|(x == 'Null')|
                                                (x == 'NA')|(x == 'na')|
                                                (x == 'none')|(x == 'unknown') else x)
data['Summary'] = data['Summary'].apply(lambda x: 'No Summary' if 
                                        (x == 'nan')|(x == 'NaN')|
                                        (x == 'N/A')|(x == '0')|
                                        (x == '')|(x == '-1')|
                                        (x == 'null')|(x == 'Null')|
                                        (x == 'NA')|(x == 'na')|
                                        (x == 'none')|(x == 'unknown') else x)
```

We would like to do a sentiment analysis, which means to understand which comments are supposedly positive and which comments are negative. Lucky for us, we have a score column.

```python
#check for score column distribution
sns.countplot(x = 'Score',data = data)
plt.title('Sentiment Distribution')
plt.show()
```

We can see from the plot above that the majority of user give score 5/5. The probability distributions it self is a little bit skewed to the left. We can say that users that gave a score 4/5 are users who's neutral. Users who gave score below 4 are users who's dissatisfied.

```python
#See what is the most used words from our data
from wordcloud import WordCloud,STOPWORDS
stpwrd = set(STOPWORDS)
stpwrd.update(['br','href']) #this is to handle should there be any HTML tags
text = ' '.join(i for i in data.Text) #this is to combine all text into one huge string.
wordcloud = WordCloud(stopwords = stpwrd, background_color='white', random_state = 42).generate(text)

plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

We can see from the wordcloud above, most of the reviews have been relatively positive if we because we see 'Love' and 'Good' appear on the wordcloud. This is if we assume our customers weren't sarcastic :)

our assumption for most of the reviews being positive is also supported by the count of 5/5 votes being relatively high.

## Classifying Sentiments

Now, we would want to see how the reviews look like if we classify customers by their score. As stated before, 5/5 score will be classified as 'good', 4/5 will be classified as neutral, below 3 score will be classified as 'bad'

```python
data['Sentiments'] = data['Score'].apply(lambda x: 1 if x == 5 else 0 if x == 4 else -1)
```

```python
print(data.Score.value_counts())
print(data.Sentiments.value_counts())
```

### Checking positive reviews

```python
postext = ' '.join(i for i in data[data.Sentiments == 1].Text) #this is to combine all text into one huge string.
poswordcloud = WordCloud(stopwords = stpwrd, background_color='white', random_state = 42).generate(postext)

plt.imshow(poswordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

### Checking neutral reviews

```python
neutext = ' '.join(i for i in data[data.Sentiments == 0].Text) #this is to combine all text into one huge string.
neuwordcloud = WordCloud(stopwords = stpwrd, background_color='white', random_state = 42).generate(neutext)

plt.imshow(neuwordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

### Checking negative reviews

```python
negtext = ' '.join(i for i in data[data.Sentiments == -1].Text) #this is to combine all text into one huge string.
negwordcloud = WordCloud(stopwords = stpwrd, background_color='white', random_state = 42).generate(negtext)

plt.imshow(negwordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

A couple of interesting things to notice: a) the more negative the review is, the more often the word "coffee" and "taste" appear, judging by the size of the words in the wordclouds. b) the more negative the review is, the rarer we can search for the word "love". c) we won't find "great" within a negative review since it will be very rare.

We can see that "good", "one","taste","make", "flavor", and "product" is a major keywords in all three classes. We will try to add those words to stopwords and rerun the wordcloud. "Amazon" is also a major keywords in all three classes but we will tend to that later.

## Rerun Wordcloud

```python
stpwrd.update(['good','one','taste','make','flavor','product'])
```

```python
newpostext = ' '.join(i for i in data[data.Sentiments == 1].Text) #this is to combine all text into one huge string.
newposwordcloud = WordCloud(width = 1600,height = 800, stopwords = stpwrd, background_color='white', random_state = 42).generate(newpostext)

fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize = (25,25))
ax1.imshow(newposwordcloud, interpolation = 'bilinear')
ax1.axis('off')
ax1.set_title('Positive Reviews Wordcloud')

newneutext = ' '.join(i for i in data[data.Sentiments == 0].Text) #this is to combine all text into one huge string.
newneuwordcloud = WordCloud(width = 1600,height = 800,stopwords = stpwrd, background_color='white', random_state = 42).generate(newneutext)

ax2.imshow(newneuwordcloud, interpolation = 'bilinear')
ax2.axis('off')
ax2.set_title('Neutral Reviews Wordcloud')

newnegtext = ' '.join(i for i in data[data.Sentiments == -1].Text) #this is to combine all text into one huge string.
newnegwordcloud = WordCloud(width = 1600,height = 800,stopwords = stpwrd, background_color='white', random_state = 42).generate(newnegtext)

ax3.imshow(newnegwordcloud, interpolation = 'bilinear')
ax3.axis('off')
ax3.set_title('Negative Reviews Wordcloud')

plt.show()
```

Our findings don't differ much from the previous findings. We can see that "Amazon" was previously noticed and now still appears in all three sentiment types. Before trying to classify the data, it's better that we add "good", "amazon", "one", "taste", "make", "flavor", and "product" to stopwords since it doesn't help differentiate the 3 sentiments that we would want to predict from the reviews.
